# Run Clean-Up

```python
btnRun = widgets.Button(description='Run Clean-Up')
out = widgets.Output()
def on_button_clicked(_):
    with out:
        file_cleanup()
        print("Running Clean-up")
        print("File Cleanup Complete")
        
btnRun.on_click(on_button_clicked)
btnRun = widgets.VBox([btnRun,out])
display(btnRun)


def file_cleanup():
    
    global ownershiplookup 
    
    #Step 1
    
    #List of new columns to be added
    new_cols = ["WFM/HCM", "Deal Reporting Type"]
    
    #add new columns to 'allclosed dataframe'
    allclosed = monthlyclosed.assign(**dict.fromkeys(new_cols, ""))
    
    
    #Step 1.2: Ci Products Closed Deal Reporting Filter. Vlookup for product Code & Platform Code
    
    #The "xls" dataframe has multiple tabs, we want to read in the "New Products Quoted" tab
    prodlookup = pd.read_excel(prodplatlookup, 'New Products Quoted')
    
    #Get "Product Code" & "Platform Code" from the prodlookup dataframe
    prodresults = monthlyproducts.merge(prodlookup, how='left', left_on='Price Book Entry: Product: Product Name', right_on='Product Name')
    
    #rename columns
    prodresults = prodresults.rename(columns={'Product':'Product Code', 'Platform':'Platform Code'})
    
    #Find and Replace values for Products
    prodresults.loc[prodresults['Product Name'].str.contains('ultipro', case=False, na=False), 'Product Name'] = 'UKG Pro'

    prodresults.loc[prodresults['Product Name'].str.contains('WFR', case=False, na=False), 'Product Name'] = 'UKG Ready'

    prodresults.loc[prodresults['Product Name'].str.contains('Workforce', case=False, na=False), 'Product Name'] = 'UKG Ready'

    prodresults.loc[prodresults['Product Name'].str.contains('Ready', case=False, na=False), 'Product Name'] = 'UKG Ready'

    prodresults.loc[prodresults['Product Name'].str.contains('WFC', case=False, na=False), 'Product Name'] = 'UKG Workforce Central'

    prodresults.loc[prodresults['Product Name'].str.contains('Legacy', case=False, na=False), 'Product Name'] = 'UKG Workforce Central'

    prodresults.loc[prodresults['Product Name'].str.contains('V6', case=False, na=False), 'Product Name'] = 'UKG Workforce Central'

    prodresults.loc[prodresults['Product Name'].str.contains('V7', case=False, na=False), 'Product Name'] = 'UKG Workforce Central'

    prodresults.loc[prodresults['Product Name'].str.contains('V8', case=False, na=False), 'Product Name'] = 'UKG Workforce Central'

    prodresults.loc[prodresults['Product Name'].str.contains('WFD', case=False, na=False), 'Product Name'] = 'UKG Dimensions'

    prodresults.loc[prodresults['Product Name'].str.contains('Dimensions', case=False, na=False), 'Product Name'] = 'UKG Dimensions'
    
    #Step 2.H
    #Create product pivot table
    prod_pivot = pd.pivot_table(prodresults, values="Opportunity Type", index="Opportunity ID", columns=["Product Code"], aggfunc=pd.Series.nunique)

    #Replace numeric values with column headers
    prod_pivot = prod_pivot.where(prod_pivot !=1, prod_pivot.columns.to_series(), axis=1)

    #Drop 'UNKOWN' column
    prod_pivot = prod_pivot.drop(columns=['UNKNOWN;'], errors='ignore')

    #Merge all columns together
    prod_pivot["Products Quoted (Formula)"] = prod_pivot[prod_pivot.columns[1:]].apply(lambda x: '' .join(x.dropna().astype(str)),axis=1)

    #Flatten prod_pivot
    prod_pivot = prod_pivot.rename_axis(None, axis=1).reset_index()
    
    
    #STEP 2.i:

    #Creat platform pivot table
    plat_pivot = pd.pivot_table(prodresults, values="Opportunity Type", index="Opportunity ID", columns=["Platform Code"], aggfunc=pd.Series.nunique)

    #Replace numeric values with column headers
    plat_pivot = plat_pivot.where(plat_pivot !=1, plat_pivot.columns.to_series(), axis=1)

    #Drop "UNKNOWN" columns
    plat_pivot = plat_pivot.drop(columns=["UNKNOWN;"], errors = 'ignore')

    #Merge all columns together
    plat_pivot["Platform Quoted (Formula)"] = plat_pivot[plat_pivot.columns[1:]].apply(lambda x: '' .join(x.dropna().astype(str)),axis=1)

    #Flatten plat_pivot
    plat_pivot = plat_pivot.rename_axis(None, axis=1).reset_index()
    

    
    #STEP 2.j

    #Merge "Products Quoted (Formula)" into allclosed
    allclosed = pd.merge(allclosed, prod_pivot[['Opportunity ID', 'Products Quoted (Formula)']], how='left', on='Opportunity ID')

    #Rename column
    allclosed = allclosed.rename(columns={"Products Quoted (Formula)":"Products Quoted"})

    #Merge "Platform Quoted (Formula)" into all Closed
    allclosed = pd.merge(allclosed, plat_pivot[['Opportunity ID', 'Platform Quoted (Formula)']], how='left', on='Opportunity ID')


    allclosed = allclosed.rename(columns={"Platform Quoted (Formula)":"Platform Quoted"})
    
    
    
    #Step 3, Clean-up 'Lead Data for Competitive Intel'
    
    #Lookup for "Product shorthand" from LEAD INTEREST TAB from .xls File
    leadlookup = pd.read_excel(prodplatlookup, "LEAD INTEREST")
    
    #get Lead Product Shorthand 1 from leadlookup dataframe
    leadprod1 = pd.merge(monthlyleads, leadlookup[['Lead Product Interest','Lead Product Shorthand']], how='left', left_on='Product Interest 1', right_on='Lead Product Interest')
    
    #rename columns
    leadprod1 = leadprod1.rename(columns={"Lead Product Shorthand":"Product VLOOKUP Shorthand 1"})

    #drop column "Lead product Interest", duplicate
    leadprod1 = leadprod1.drop(columns=['Lead Product Interest'])
    
    #get Lead Product Shorthand 2 from leadlookup dataframe
    leadprod2 = pd.merge(monthlyleads, leadlookup[['Lead Product Interest', 'Lead Product Shorthand']], how='left', left_on='Product Interest 2', right_on='Lead Product Interest')
    
    #rename columns
    leadprod2 = leadprod2.rename(columns={"Lead Product Shorthand":"Product VLOOKUP Shorthand 2"})

    #drop column "Lead product Interest", duplicate
    leadprod2 = leadprod2.drop(columns=['Lead Product Interest'])

    #combine both dataframes to get both "Product VLOOKUP Shorthand" fields
    leadresults = pd.merge(leadprod1, leadprod2[['Opportunity ID', 'Product VLOOKUP Shorthand 2']], how='left', on='Opportunity ID')
    
    #Combine VLOOKUP 1 Shorthand & VLOOKUP 2 Shorthand fields
    leadresults['COMBO FORMULA'] = leadresults['Product VLOOKUP Shorthand 1'].fillna(leadresults['Product VLOOKUP Shorthand 2']) + leadresults['Product VLOOKUP Shorthand 2'].fillna(leadresults['Product VLOOKUP Shorthand 1'])
    
    
    #Merge 'COMBO FORMULA' column into 'allclosed' dataframe

    allclosed = pd.merge(allclosed, leadresults[['Opportunity ID', 'COMBO FORMULA']], how='left', on='Opportunity ID')

    allclosed = allclosed.rename(columns={"COMBO FORMULA":"Product Interest on Lead"})
    
    allclosed.to_csv("allclosed.csv", index=False)
    
    
    #Step 4, 'Product Mentioned in OppName'

    #Creat 'oppname' DataFrame
    oppname = allclosed[['Opportunity ID', 'Opportunity Name']].copy()


    #Create product lookup columns

    booldict = {True:'TRUE', False:'FALSE'}

    ACT = ['Activity','ACTVTY','Activities']
    oppname['ACT'] = oppname['Opportunity Name'].str.contains('|'.join(ACT))
    oppname['ACT'] = oppname['ACT'].map(booldict)
    oppname.loc[oppname['ACT'] == 'TRUE', 'ACT'] = 'ACT;'
    oppname.loc[oppname['ACT'] == 'FALSE', 'ACT'] = ''

    Analytics = ['WFAN','Analy','-WAH','Analytic','WAH']
    oppname['Analytics'] = oppname['Opportunity Name'].str.contains('|'.join(Analytics))
    oppname['Analytics'] = oppname['Analytics'].map(booldict)
    oppname.loc[oppname['Analytics'] == 'TRUE', 'Analytics'] = 'Analytics;'
    oppname.loc[oppname['Analytics'] == 'FALSE', 'Analytics'] = ''

    ATK = ['Attestation','ATK','Attest']
    oppname['ATK'] = oppname['Opportunity Name'].str.contains('|'.join(ATK))
    oppname['ATK'] = oppname['ATK'].map(booldict)
    oppname.loc[oppname['ATK'] == 'TRUE', 'ATK'] = 'ATK;'
    oppname.loc[oppname['ATK'] == 'FALSE', 'ATK'] = ''

    ATS = ['ATS','Talent',' ATS','Hiring','ATS/','Recruit','Applicant Track']
    oppname['ATS'] = oppname['Opportunity Name'].str.contains('|'.join(ATS))
    oppname['ATS'] = oppname['ATS'].map(booldict)
    oppname.loc[oppname['ATS'] == 'TRUE', 'ATS'] = 'ATS;'
    oppname.loc[oppname['ATS'] == 'FALSE', 'ATS'] = ''

    EDU = ['Education','EDU Points','K Pass','Knowledge Pass','KnowledgePass','Kpass','K-pass','Training','User Adoption']
    oppname['EDU'] = oppname['Opportunity Name'].str.contains('|'.join(EDU))
    oppname['EDU'] = oppname['EDU'].map(booldict)
    oppname.loc[oppname['EDU'] == 'TRUE', 'EDU'] = 'EDU;'
    oppname.loc[oppname['EDU'] == 'FALSE', 'EDU'] = ''

    EEMGR = ['0 lice','0 Manager','License Add','0 MGR','Omgr','5 lice','5 Manager','5 MGR','5mgr','Admin','cap add','Cap Inc','capacity','CapInc','EE Cap','EE Lic','Employee Lice','Emps','ESS','Manager Cap','Manager Lic','Managers','Mgr Cap','MGR Lic','MGRs','WF EE','add lic','WF Emp','WF MGR','WFEE','WFMGR','ty increase','addtl lic','Emp&Mgr']
    oppname['EE/MGR'] = oppname['Opportunity Name'].str.contains('|'.join(EEMGR))
    oppname['EE/MGR'] = oppname['EE/MGR'].map(booldict)
    oppname.loc[oppname['EE/MGR'] == 'TRUE', 'EE/MGR'] = 'EE/MGR;'
    oppname.loc[oppname['EE/MGR'] == 'FALSE', 'EE/MGR'] = ''

    EZCall = ['EZ Call','EZCall','Physician Sch']
    oppname['EZCall'] = oppname['Opportunity Name'].str.contains('|'.join(EZCall))
    oppname['EZCall'] = oppname['EZCall'].map(booldict)
    oppname.loc[oppname['EZCall'] == 'TRUE', 'EZCall'] = 'EZCall;'
    oppname.loc[oppname['EZCall'] == 'FALSE', 'EZCall'] = ''

    Hosting = ['Hosting','Hosted','KPC']
    oppname['Hosting'] = oppname['Opportunity Name'].str.contains('|'.join(Hosting))
    oppname['Hosting'] = oppname['Hosting'].map(booldict)
    oppname.loc[oppname['Hosting'] == 'TRUE', 'Hosting'] = 'Hosting;'
    oppname.loc[oppname['Hosting'] == 'FALSE', 'Hosting'] = ''

    HR = ['HR;','/HR','HRIS','HR ',';HR','FS ','R Suite','C Suite','HR,',',HR','Performance M','HR/','& HR','Full Ste','Full Suite','FullSuite',' HR','HCM',' FS','HR-','-HR','Onboarding','HR&','Compensation','&HR','Human Resources','WFC Full','Full-Suite','HR &','WFR Full','HRMS']
    oppname['HR'] = oppname['Opportunity Name'].str.contains('|'.join(HR))
    oppname['HR'] = oppname['HR'].map(booldict)
    oppname.loc[oppname['HR'] == 'TRUE', 'HR'] = 'HR;'
    oppname.loc[oppname['HR'] == 'FALSE', 'HR'] = ''

    HW = ['4500s',"4500's",'Bio Add','POE Kit','Badge','Bio Reader','bio unit','biometric','Clock','Hardware','HW','In Touch','InTouch','In-Touch','4500','Prox Kit','TID Unit','Reader','rental','Scanner','Terminal','Touch ID','Trade']
    oppname['HW'] = oppname['Opportunity Name'].str.contains('|'.join(HW))
    oppname['HW'] = oppname['HW'].map(booldict)
    oppname.loc[oppname['HW'] == 'TRUE', 'HW'] = 'HW;'
    oppname.loc[oppname['HW'] == 'FALSE', 'HW'] = ''

    iSeries = ['i-Series','iSeries','i Series']
    oppname['iSeries'] = oppname['Opportunity Name'].str.contains('|'.join(iSeries))
    oppname['iSeries'] = oppname['iSeries'].map(booldict)
    oppname.loc[oppname['iSeries'] == 'TRUE','iSeries'] = 'iSeries;'
    oppname.loc[oppname['iSeries'] == 'FALSE','iSeries'] = ''

    KSS = ['acaplug','acareport','KSS','ACA','(ACA)',',ACA','ACA Report',', ACA','Aca Plug','; ACA',';ACA','aca ,','ACA ;','ACA;','Ft/PT','/ACA','ACA/','TK ACA','&ACA','& ACA','WFR ACA','ACA Web', 'ACA opp','ACA proj','- ACA','-ACA','ACA &']
    oppname['KSS'] = oppname['Opportunity Name'].str.contains('|'.join(KSS))
    oppname['KSS'] = oppname['KSS'].map(booldict)
    oppname.loc[oppname['KSS'] == 'TRUE', 'KSS'] = 'KSS;'
    oppname.loc[oppname['KSS'] == 'FALSE', 'KSS'] = ''

    Mobile = ['Tablet','ipad','Mobile']
    oppname['Mobile'] = oppname['Opportunity Name'].str.contains('|'.join(Mobile))
    oppname['Mobile'] = oppname['Mobile'].map(booldict)
    oppname.loc[oppname['Mobile'] == 'TRUE', 'Mobile'] = 'Mobile/Tablet;'
    oppname.loc[oppname['Mobile'] == 'FALSE', 'Mobile'] = ''

    PR = ['PR','PY','Full-Suite','&PR',',PR','FS','Full Ste','Full Suite','HRMS','and Pay','TK Pay','FullSuite','HCM','HR,','HRIS','Payroll','R Suite','C Suite','WFR Full','WFC Full','Ready Full','Central Full','PR-','PR&','PR,','PR/','PR;']
    oppname['PR'] = oppname['Opportunity Name'].str.contains('|'.join(PR))
    oppname['PR'] = oppname['PR'].map(booldict)
    oppname.loc[oppname['PR'] == 'TRUE', 'PR'] = 'PR;'
    oppname.loc[oppname['PR'] == 'FALSE','PR'] = ''

    SaaS = ['Cloud','SaaS']
    oppname['SaaS'] = oppname['Opportunity Name'].str.contains('|'.join(SaaS))
    oppname['SaaS'] = oppname['SaaS'].map(booldict)
    oppname.loc[oppname['SaaS'] == 'TRUE', 'SaaS'] = 'SaaS;'
    oppname.loc[oppname['SaaS'] == 'FALSE', 'SaaS'] = ''

    SCHED = ['Extension','Forecast','OptiLink','QuickFill','SCHED','SMS','WFS','Workload Manager','Workload Mgr','Adv SCH','WFC SCH','Target Intel']
    oppname['SCHED'] = oppname['Opportunity Name'].str.contains('|'.join(SCHED))
    oppname['SCHED'] = oppname['SCHED'].map(booldict)
    oppname.loc[oppname['SCHED'] == 'TRUE', 'SCHED'] = 'SCHED;'
    oppname.loc[oppname['SCHED'] == 'FALSE', 'SCHED'] = ''

    Services = ['add on','add user','license add','Invoice Issue','location add','Additional','change order','work order','Addon','Add-On','Assess','Build','cap add','Cap Inc','capacity','Config','Consult','Custom Re','Customiz','deduc','expansion','export','Fixed Fee','hours','Implement','import','Install','Integrate','Interface','KPS','Licenses','Meal Break','new location','new site','new store','Optimiza','PBJ','policies','policy','Prof Serv','PS Only','Report','Rules','scope','scoping','Service','Services Only','setup','SOW','Svcs','401k','401(k)','workrule','work rule','serv only','single sign','upgrade','Upgrade','upgrd','v8','WFC Upg']
    oppname['Services'] = oppname['Opportunity Name'].str.contains('|'.join(Services))
    oppname['Services'] = oppname['Services'].map(booldict)
    oppname.loc[oppname['Services'] == 'TRUE', 'Services'] = 'Services;'
    oppname.loc[oppname['Services'] == 'FALSE', 'Services'] = ''

    TK = ['/ TA','/TA','-FS','T/A','t/I','Full ste','Full Suite','FullSuite','Full-Suite','IVR','keeper','T&A','T&L','T/A','R Suite','C Suite','Time','TK','TLM','TTE','WF Management','WFC Full','WFCFS','WFCFULL','WFM','WFR FS','WFR Full','WFRFS']
    oppname['TK'] = oppname['Opportunity Name'].str.contains('|'.join(TK))
    oppname['TK'] = oppname['TK'].map(booldict)
    oppname.loc[oppname['TK'] == 'TRUE', 'TK'] = 'TK;'
    oppname.loc[oppname['TK'] == 'FALSE','TK'] = ''

    TKC = ['TKC','VISIONW','Web TA','WebTA']
    oppname['TKC'] = oppname['Opportunity Name'].str.contains('|'.join(TKC))
    oppname['TKC'] = oppname['TKC'].map(booldict)
    oppname.loc[oppname['TKC'] == 'TRUE', 'TKC'] = 'TKC/VISIONW/WebTA/SMART;'
    oppname.loc[oppname['TKC'] == 'FALSE', 'TKC'] = ''

    TSG = ['bidding','NEW TS','TeleStaff','TS Upgrade','TSG','WFTS','auctions','institution focus','gateway to wfc']
    oppname['TSG'] = oppname['Opportunity Name'].str.contains('|'.join(TSG))
    oppname['TSG'] = oppname['TSG'].map(booldict)
    oppname.loc[oppname['TSG'] == 'TRUE', 'TSG'] = 'TSG;'
    oppname.loc[oppname['TSG'] == 'FALSE','TSG'] = ''

    WAM = [' ACC',' WAM','Absence','ACC-','Accrual','Attend','FLSA','Leave','WAM ']
    oppname['WAM'] = oppname['Opportunity Name'].str.contains('|'.join(WAM))
    oppname['WAM'] = oppname['WAM'].map(booldict)
    oppname.loc[oppname['WAM'] == 'TRUE', 'WAM'] = 'WAM;'
    oppname.loc[oppname['WAM'] == 'FALSE', 'WAM'] = ''

    WIM = ['benefit feed','export','gration','import','integrat','interface','WIM','WIM,']
    oppname['WIM'] = oppname['Opportunity Name'].str.contains('|'.join(WIM))
    oppname['WIM'] = oppname['WIM'].map(booldict)
    oppname.loc[oppname['WIM'] == 'TRUE', 'WIM'] = 'WIM;'
    oppname.loc[oppname['WIM'] == 'FALSE', 'WIM'] =''

    WFReady = ['flip to ready','WF-Ready','WFR','WF Ready','Ready C','Workforce Ready','WorkforceR']
    oppname['WFReady'] = oppname['Opportunity Name'].str.contains('|'.join(WFReady))
    oppname['WFReady'] = oppname['WFReady'].map(booldict)
    oppname.loc[oppname['WFReady'] == 'TRUE', 'WFReady'] = 'WF Ready;'
    oppname.loc[oppname['WFReady'] == 'FALSE', 'WFReady'] = ''

    WFC = ['Activ','Analy','Extension','Upgrade','v8','v 8','Onprem','On-prem','Tablet','WAM','WF Central','WFC','WIM','Workforce Central']
    oppname['WFC'] = oppname['Opportunity Name'].str.contains('|'.join(WFC))
    oppname['WFC'] = oppname['WFC'].map(booldict)
    oppname.loc[oppname['WFC'] == 'TRUE', 'WFC'] = 'WFC On-Premise;'
    oppname.loc[oppname['WFC'] == 'FALSE','WFC'] = ''

    WFCsa = ['Cloud','Host','Hosted','KPC','Migrate','SAAS','SMB']
    oppname['WFCsa'] = oppname['Opportunity Name'].str.contains('|'.join(WFCsa))
    oppname['WFCsa'] = oppname['WFCsa'].map(booldict)
    oppname.loc[oppname['WFCsa'] == 'TRUE', 'WFCsa'] = 'WFC-SaaS;'
    oppname.loc[oppname['WFCsa'] == 'FALSE', 'WFCsa'] = ''

    FMSI = ['FMSI']
    oppname['FMSI'] = oppname['Opportunity Name'].str.contains('|'.join(FMSI))
    oppname['FMSI'] = oppname['FMSI'].map(booldict)
    oppname.loc[oppname['FMSI'] == 'TRUE','FMSI'] = 'FMSI;'
    oppname.loc[oppname['FMSI'] == 'FALSE','FMSI'] = ''

    Dimensions = ['Dimensions','WFD','WFDimensions','Workforce Dimensions']
    oppname['Dimensions'] = oppname['Opportunity Name'].str.contains('|'.join(Dimensions))
    oppname['Dimensions'] = oppname['Dimensions'].map(booldict)
    oppname.loc[oppname['Dimensions'] == 'TRUE', 'Dimensions'] = 'Workforce Dimensions;'
    oppname.loc[oppname['Dimensions'] == 'FALSE', 'Dimensions'] = ''


    
    
    #Combine 'Products in OppName'
    oppname['Product Mentioned in Opp Name'] = oppname['ACT']+oppname['Analytics']+oppname['ATK']+oppname['ATS']+oppname['EDU']+oppname['EE/MGR']+oppname['EZCall']+oppname['Hosting']+oppname['HR']+oppname['HW']+oppname['iSeries']+oppname['KSS']+oppname['Mobile']+oppname['PR']+oppname['SaaS']+oppname['SCHED']+oppname['Services']+oppname['TK']+oppname['TKC']+oppname['TSG']+oppname['WAM']+oppname['WIM']+oppname['FMSI']

    oppname['Platform in Opp Name'] = oppname['WFReady']+oppname['iSeries']+oppname['WFC']+oppname['WFCsa']+oppname['Dimensions']

    oppname = oppname.rename(columns={'Opportunity ID':'Opp ID'})

    #merge 'Products mentioned in Opp Name' to allclosed
    allclosed = pd.merge(allclosed, oppname[['Opp ID', 'Product Mentioned in Opp Name']], how='left', left_on='Opportunity ID', right_on='Opp ID')

    allclosed = allclosed.drop(columns=['Opp ID'])

    #merge 'Platform in Opp Name' to allclosed
    allclosed = pd.merge(allclosed, oppname[['Opp ID', 'Platform in Opp Name']], how='left', left_on='Opportunity ID',right_on='Opp ID')

    allclosed = allclosed.drop(columns=['Opp ID'])

    
    
    #Step 7, complete competitor information in 'allclosed'
    
    #Merge allclosed with Master Competitor file to get "Competitor Parent (Revised)"
    allclosed = pd.merge(allclosed, competitorlookup[['Competitor in SF', 'Competitor Parent (Revised)']], how='left', left_on='Competitor', right_on='Competitor in SF')

    allclosed = allclosed.drop(columns=['Competitor in SF'])

    #Merge allclosed with Master Competitor file to get "Competitor (Revised)"
    allclosed = pd.merge(allclosed, competitorlookup[['Competitor in SF', 'Competitor (Revised)']], how='left', left_on='Competitor', right_on='Competitor in SF')

    allclosed = allclosed.drop(columns=['Competitor in SF'])

    #Merge allclosed with Master Competitor file to get "Competitor Class"
    allclosed = pd.merge(allclosed, competitorlookup[['Competitor in SF', 'Competitor Class ']], how='left', left_on='Competitor', right_on='Competitor in SF')

    allclosed = allclosed.drop(columns=['Competitor in SF'])

    allclosed = allclosed.rename(columns={'Competitor':'Competitive?'})
    
    #Replace "None" vales in Competitor column with "Non-Competitive"
    allclosed['Competitive?'] = allclosed['Competitive?'].replace('None','Non-Competitive')
    
    #Replace values for 'Competitive'
    allclosed.loc[(allclosed['Competitive?'] != 'Non-Competitive') & (allclosed['Competitive?'] != 'Unknown'), 'Competitive?'] ='Competitive'
    
    
    #Step 8, Complete Products (Revised) column

    allclosed['Products (Revised)'] = allclosed['Products Quoted'].fillna(allclosed['Product Mentioned in Opp Name'])
    allclosed['Products (Revised)'] = allclosed['Products (Revised)'].fillna(allclosed['Product Interest on Lead'])
    
    
    #Step 9, Complete Platform (Revised) column

    allclosed['Platform (Revised)'] = allclosed['Platform Quoted'].fillna(allclosed['Platform in Opp Name'])
    
    
    #Step 12, Rename "Age" column to "Days Open"
    allclosed = allclosed.rename(columns={'Age':'Days Open'})
    
    
    #Step 13, Determine Prior Sales Stage
    
    #create sales stage dataframe
    salesstage = {'Prior Stage': ['0 - Qualification', '1 - Discovery', '1 - Prepare', '2 - Develop Solution', '3 - Proposal', '4 - Deliver Quote', '4 - Vendor Selected', '5 - Contracts'], 'Prior Sales Stage Group': ['Stage: 0-2', 'Stage: 0-2', 'Stage: 0-2','Stage: 0-2', 'Stage: 3-5', 'Stage: 3-5', 'Stage: 3-5', 'Stage: 3-5']}

    priorstage = pd.DataFrame(data=salesstage)

    #merge 'priorstage' dataframe with 'allclosed'

    allclosed = pd.merge(allclosed, priorstage, how='left', left_on='Prior Stage', right_on='Prior Stage')
    
    
    #Step 14, Determine SIC Segment
    
    #merge 'allclosed' wit 'siclookup' to get SIC Segment
    allclosed = pd.merge(allclosed, siclookup[['Revised SIC Code', 'Revised PI Segment']], how='left', left_on='SIC Code', right_on='Revised SIC Code')

    #drop column
    allclosed = allclosed.drop(columns=['Revised SIC Code', 'Sales Position Segment'])
    
    #rename column
    allclosed = allclosed.rename(columns={'Revised PI Segment':'SIC Segment'})
    
    
    #Step 15, Determine "Forecast ARR"
    
    #Drop all USD columns
    allclosed = allclosed.drop([x for x in allclosed if x.endswith('Currency')], 1)
    
    #rename columns
    allclosed.columns.values[36] = 'Total ARR Equiv and ARR (converted)'
    allclosed.columns.values[37] = 'Grand Total ARR Equiv and ARR (converted)'
    
    
    #Remove '0' from ARR columns
    allclosed.loc[allclosed['Forecast ARR (converted)'] == 0, 'Forecast ARR (converted)'] = ''
    allclosed.loc[allclosed['Total ARR Equiv and ARR (converted)'] == 0, 'Total ARR Equiv and ARR (converted)'] = ''
    allclosed.loc[allclosed['Grand Total ARR Equiv and ARR (converted)'] == 0, 'Grand Total ARR Equiv and ARR (converted)'] = ''
    allclosed.loc[allclosed['Quote - Total ARR Equiv (converted)'] == 0, 'Quote - Total ARR Equiv (converted)'] = ''
    allclosed.loc[allclosed['Quote - Grand Total ARR Equiv and ARR (converted)'] == 0, 'Quote - Grand Total ARR Equiv and ARR (converted)'] = ''
    
    
    #Fill in 'Forecast ARR' columns
    allclosed['Forecast ARR (converted)'] = allclosed['Forecast ARR (converted)'].fillna(allclosed['Total ARR Equiv and ARR (converted)'])
    allclosed['Forecast ARR (converted)'] = allclosed['Forecast ARR (converted)'].fillna(allclosed['Grand Total ARR Equiv and ARR (converted)'])
    allclosed['Forecast ARR (converted)'] = allclosed['Forecast ARR (converted)'].fillna(allclosed['Quote - Total ARR Equiv (converted)'])
    allclosed['Forecast ARR (converted)'] = allclosed['Forecast ARR (converted)'].fillna(allclosed['Quote - Grand Total ARR Equiv and ARR (converted)'])

    
    
    #Step 16, Determine EE Size
    
    #Create 'Hourly+Salary' field
    leadresults['Hourly+Salary'] = leadresults['# Salary Employees'] + leadresults['# Hourly Employees']
    leadresults['Hourly+Salary'] = leadresults['Hourly+Salary'].fillna(leadresults['# Salary Employees'])
    leadresults['Hourly+Salary'] = leadresults['Hourly+Salary'].fillna(leadresults['# Hourly Employees'])
    
    #Create EE Range lookup table
    rangeconverter = {'Lead EE Range': ['< 100', '100-599', '600-2499', '2500+'], 'Lead EE Revise': ['50', '250', '1500', '2501']}

    eeconverter = pd.DataFrame(data=rangeconverter)
    
    #Merge leadresults with eeconverter to create 'EE on Lead' field
    leadresults = pd.merge(leadresults, eeconverter, how='left', left_on='Total Employees', right_on='Lead EE Range')

    #clean-up columns
    leadresults= leadresults.drop(columns=['Lead EE Range'])
    leadresults = leadresults.rename(columns={'Lead EE Revise':'EE on Lead'})
    
    #replace 'Unknown' values with blanks
    leadresults['Total Employees'].replace({'Unknown':None}, inplace=True)
    
    #fill 'EE on Lead' column with Total Employees & Hourly+Salary columns
    leadresults['EE on Lead'] = leadresults['Total Employees'].fillna(leadresults['Hourly+Salary'])
    
    
    
    #Step 17, Final EE count in allclosed dataframe
    
    #rename EE columns
    allclosed = allclosed.rename(columns={'# of Employees':'EE on Opp'})
    allclosed = allclosed.rename(columns={'Employees Company Total (UKG)':'EE on Account'})
    
    
    #merge 'EE on Lead' column from leadresults to allclosed
    allclosed = pd.merge(allclosed, leadresults[['Opportunity ID', 'EE on Lead']], how='left', left_on='Opportunity ID', right_on='Opportunity ID')
    
    #create 'EE Size (Revised)' column 
    allclosed['EE Size (Revised)'] = allclosed['EE on Opp'].fillna(allclosed['EE on Lead'])
    allclosed['EE Size (Revised)'] = allclosed['EE Size (Revised)'].fillna(allclosed['EE on Account'])
    
    
    
    #Step 18, Determine Ownership
    
    #clean up ownership VLOOKUP file
    new_header = ownershiplookup.iloc[0]
    ownershiplookup = ownershiplookup[1:]
    ownershiplookup.columns = new_header
    ownershiplookup = ownershiplookup.iloc[:,:5]
    
    #Convert "Territory IDs in both dataframes to string types"
    ownershiplookup['TerritoryID'] = ownershiplookup['TerritoryID'].astype(str)
    allclosed['Territory ID'] = allclosed['Territory ID'].astype(str)
    
    #Rename column headers
    ownershiplookup = ownershiplookup.rename(columns={'Vertical':'Vertical (Revised)', 'Sales Team':'Sales Team (Revised)', 'Region':'Region (Revised)'})
    
    #Merge ownership lookup file with allclosed dataframe
    allclosed = pd.merge(allclosed, ownershiplookup, how='left', left_on='Territory ID', right_on='TerritoryID')

    #Drop territoryID column 
    allclosed = allclosed.drop(columns=['TerritoryID'])
    
    #Clean up Territory ID column
    allclosed['Territory ID'] = allclosed['Territory ID'].fillna('')
    
    
    #Step 19, Insert 'FY QTR field'

    allclosed['FY QTR'] = allclosed['Fiscal Year'].astype(str) + " " + allclosed['Fiscal Quarter'].astype(str)
    
    
    #Re-Index Columns
    allclosed = allclosed.reindex(columns= ['Opportunity ID','Opportunity Owner: Manager','Opportunity Owner','Opportunity Owner Email',
                                            'Vertical (Revised)','Sales Team (Revised)','Region (Revised)','Territory ID','Industry Buckets',
                                            'Area','D&B Company Name','Account Name','EDW Customer Category','UKG Dom Vertical','Business Unit',
                                            'Business Unit Intl','Opportunity Account Type','Account Type','SIC Code','SIC Segment','Opportunity Name',
                                            'Created Date','Closed Date','Fiscal Year','Fiscal Quarter','Fiscal Month','FY QTR','Days Open',
                                            'Sales Stage','Reason Lost','Reason Lost Detail','Reason Won/Lost Comment','Prior Sales Stage Group',
                                            'Prior Stage','Last Sales Stage Change Date','Competitor Parent (Revised)','Competitor (Revised)',
                                            'Competitor','Competitor Class','Competitor Other Details','Products (Revised)','Products Quoted',
                                            'Product Interest on Lead','Product Mentioned in Opp Name','Platform (Revised)','Platform Quoted',
                                            'Platform in Opp Name','WFM/HCM','Deal Reporting Type','Solution Category','EE on Lead','EE Size (Revised)',
                                            'EE on Opp','EE on Account','Forecast ARR (converted)','Total ARR Equiv and ARR (converted)',
                                            'Grand Total ARR Equiv and ARR (converted)','Quote - Total ARR Equiv (converted)', 'Quote Grand Total ARR Equiv and ARR (converted)',
                                            'Opportunity Type','Offering','Cloud Offering','Comments','SR Comments','SM Comments','Oracle Account Number',
                                            'DNB Global Ultimate Duns','DNB Domestic Ultimate DUNS','Line Items','Lead','Primary Contact','Primary Contact.1',
                                            'Contact: Job Title','Contact: Phone','Contact: Email','Phone','Notes'
                                           ])
    
    
    
    allclosed.to_csv("MonthlyWinLoss.csv", index=False)
    
   
HTML('''<script>code_show=true; function code_toggle() { if (code_show){ $('div.input').hide();
} else { $('div.input').show(); } code_show = !code_show} $( document ).ready(code_toggle);
</script>The raw code for this IPython notebook is by default hidden for easier reading.
To toggle on/off the raw code, click <a href="javascript:code_toggle()">here</a>.<br><br><br><br><br><br>''')
    
    
```
